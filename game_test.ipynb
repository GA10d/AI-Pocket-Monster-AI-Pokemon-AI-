```python
import pygame
from pygame.locals import *
import sys
import image
import button
```

```python
def draw_text(string, smooth = True, color = (0,0,0), pos = (0,0),font_size = 24):
    global screen
    font = pygame.font.Font("font\HYPixel11pxU-2.ttf", font_size)
    string = font.render(string, smooth, color)
    screen.blit(string, pos)
```

```python
# 初始化Pygame
pygame.init()
pygame.display.set_caption('Display JPEG Image')

# 设置屏幕大小
screen_width = 800
screen_height = 600
screen_size = (screen_width, screen_height)
screen = pygame.display.set_mode(screen_size)
screen.fill((255,255,255))
# 设置按钮大小
button_image = pygame.image.load('image/button.png')
button_height = button_image.get_width()
button_width = button_image.get_height()
button_size = (button_width, button_height)
button_offset = 245
#
button_11 = button.Button(image = button_image, scale = 0.5, pos = (0,260), num = 1)
button_12 = button.Button(image = button_image, scale = 0.5, pos = (0,360), num = 2)
button_13 = button.Button(image = button_image, scale = 0.5, pos = (0,460), num = 3)

button_21 = button.Button(image = button_image, scale = 0.5, pos = (screen_width-button_offset,260), num = 4)
button_22 = button.Button(image = button_image, scale = 0.5, pos = (screen_width-button_offset,360), num = 5)
button_23 = button.Button(image = button_image, scale = 0.5, pos = (screen_width-button_offset,460), num = 6)
#导入技能图片
icon_scale = 0.05
icon_offsetx = 245
icon_offsety = 10

skill_11 = pygame.image.load('playerA/playerA_skill_icon1.jpeg')
icon_11 = button.Button(image = skill_11, scale = icon_scale, pos = (0,270+icon_offsety), num = 1)
skill_12 = pygame.image.load('playerA/playerA_skill_icon2.jpeg')
icon_12 = button.Button(image = skill_12, scale = icon_scale, pos = (0,370+icon_offsety), num = 2)
skill_13 = pygame.image.load('playerA/playerA_skill_icon3.jpeg')
icon_13 = button.Button(image = skill_13, scale = icon_scale, pos = (0,470+icon_offsety), num = 3)

skill_21 = pygame.image.load('playerB/playerB_skill_icon1.jpeg')
icon_21 = button.Button(image = skill_21, scale = icon_scale, pos = (screen_width-icon_offsetx,270+icon_offsety), num = 4)
skill_22 = pygame.image.load('playerB/playerB_skill_icon2.jpeg')
icon_22 = button.Button(image = skill_22, scale = icon_scale, pos = (screen_width-icon_offsetx,370+icon_offsety), num = 5)
skill_23 = pygame.image.load('playerB/playerB_skill_icon3.jpeg')
icon_23 = button.Button(image = skill_23, scale = icon_scale, pos = (screen_width-icon_offsetx,470+icon_offsety), num = 6)



# 设置VS
vs_image = pygame.image.load('image/vs.png')
vs_height = button_image.get_width()
vs_width = button_image.get_height()

vs = button.Button(image = vs_image, scale = 0.1, pos = ((screen_width-vs_width)/2-20,0), num = 1)

# 设置怪物大小
monster_width = 256
monster_height = 256
monster_size = (monster_width, monster_height)


#image = pygame.image.load('playerB/playerB_image1.jpeg')
img_MonsterA = image.Image('playerA/playerA_image1.jpeg',(0,0),size = monster_size)
img_MonsterB = image.Image('playerB/playerB_image1.jpeg',(screen_width-monster_width,0),size = monster_size)



# 游戏主循环
running = True
while running:
    # 事件处理
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # 获取鼠标点击位置
            if event.button == 2:
                running = False
            mouse_pos = pygame.mouse.get_pos()
            button_11.update(mouse_pos)
            print(mouse_pos, button_11.clicked)
            print(button_11.rect)

    # 将图片绘制到屏幕上
    img_MonsterA.draw(screen)
    img_MonsterB.draw(screen)
    vs.draw(screen)

    # 绘制按钮
    button_11.draw(screen)
    button_12.draw(screen)
    button_13.draw(screen)
    button_21.draw(screen)
    button_22.draw(screen)
    button_23.draw(screen)

    #绘制技能图标
    icon_11.draw(screen)
    icon_12.draw(screen)
    icon_13.draw(screen)
    icon_21.draw(screen)
    icon_22.draw(screen)
    icon_23.draw(screen)


    # 显示文字
    line_1 = 290
    line_2 = 390
    line_3 = 490

    row_1 = 75
    row_2 = 630

    draw_text("技能演示", pos = (row_1,line_1))
    draw_text("技能演示", pos = (row_1,line_2))
    draw_text("技能演示", pos = (row_1,line_3))
    draw_text("技能演示", pos = (row_2,line_1))
    draw_text("技能演示", pos = (row_2,line_2))
    draw_text("技能演示", pos = (row_2,line_3))

    # 更新屏幕显示
    pygame.display.flip()

# 退出Pygame
pygame.quit()
sys.exit()
```
